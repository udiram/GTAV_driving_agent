```python
import numpy as np
import pandas as pd
from collections import Counter
from random import shuffle
```

```python
train_data = np.load('data/training_data.npy', allow_pickle=True)

df = pd.DataFrame(train_data)
print(Counter(df[1].apply(str)))
df.head()
```

```python
W, S, A, D, WA, WD, SA, SD, NOKEY = [], [], [], [], [], [], [], [], []
shuffle(train_data)
```

```python
for data in train_data:
    img = data[0]
    choice = data[1]
    print(choice)
    if choice == [1, 0, 0, 0, 0, 0, 0, 0, 0]:
        W.append([img, choice])
    elif choice == [0, 1, 0, 0, 0, 0, 0, 0, 0]:
        S.append([img, choice])
    elif choice == [0, 0, 1, 0, 0, 0, 0, 0, 0]:
        A.append([img, choice])
    elif choice == [0, 0, 0, 1, 0, 0, 0, 0, 0]:
        D.append([img, choice])
    elif choice == [0, 0, 0, 0, 1, 0, 0, 0, 0]:
        WA.append([img, choice])
    elif choice == [0, 0, 0, 0, 0, 1, 0, 0, 0]:
        WD.append([img, choice])
    elif choice == [0, 0, 0, 0, 0, 0, 1, 0, 0]:
        SA.append([img, choice])
    elif choice == [0, 0, 0, 0, 0, 0, 0, 1, 0]:
        SD.append([img, choice])
    elif choice == [0, 0, 0, 0, 0, 0, 0, 0, 1]:
        NOKEY.append([img, choice])
    else:
        print('no matches')
```

```python
#BALANCE THIS DATASET
```

```python
W = W[:len(S)][:len(A)][:len(D)][:len(WA)][:len(WD)][:len(SA)][:len(SD)][:len(NOKEY)]
S = S[:len(W)][:len(A)][:len(D)][:len(WA)][:len(WD)][:len(SA)][:len(SD)][:len(NOKEY)]
A = A[:len(W)][:len(S)][:len(D)][:len(WA)][:len(WD)][:len(SA)][:len(SD)][:len(NOKEY)]
D = D[:len(W)][:len(S)][:len(A)][:len(WA)][:len(WD)][:len(SA)][:len(SD)][:len(NOKEY)]
WA = WA[:len(W)][:len(S)][:len(A)][:len(D)][:len(WD)][:len(SA)][:len(SD)][:len(NOKEY)]
WD = WD[:len(W)][:len(S)][:len(A)][:len(D)][:len(WA)][:len(SA)][:len(SD)][:len(NOKEY)]
SA = SA[:len(W)][:len(S)][:len(A)][:len(D)][:len(WA)][:len(WD)][:len(SD)][:len(NOKEY)]
SD = SD[:len(W)][:len(S)][:len(A)][:len(D)][:len(WA)][:len(WD)][:len(SA)][:len(NOKEY)]
NOKEY = NOKEY[:len(W)][:len(S)][:len(A)][:len(D)][:len(WA)][:len(WD)][:len(SA)][:len(SD)]
```

```python
final_data = W + S + A + D + WA + WD + SA + SD + NOKEY
shuffle(final_data)
print(len(final_data))
```

```python
np.save('data/training_data_balanced.npy', final_data)
```
